```python
#Import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd

#URL of the CCNY Fall 2021 academic calender
url = 'https://www.ccny.cuny.edu/registrar/fall'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extracts the response as html: html_doc
html_doc = r.text

# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc, 'html.parser')

# Get Calendar text: guido_text
#calendar_text = soup.get_text()

# Print Calendar text to the shell
#print(calendar_text)

#Find all 'td' tags
calendar_info = soup.find_all('td')

#Extract data from each row in the table
rows = []
for i in range(0, len(calendar_info), 3):
  date = calendar_info[i].get_text(strip=True)
  dow = calendar_info[i + 1].get_text(strip=True)
  text = calendar_info[i + 2].get_text(strip=True).replace('\n', ' ').replace('\t', ' ')
  rows.append([date, dow, text])

#Create DataFrame
df = pd.DataFrame(rows, columns=['Date', 'DOW', 'Text'])

#Set the date column as the index
df.set_index('Date', inplace=True)

#Print the DataFrame
display(df)
```
